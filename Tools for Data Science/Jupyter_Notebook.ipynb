### Obtain the mole using a measured NaCl mass in grams

```python
def get_NaCl_mole(mass_in_grams):
    Na_atomic_weight = 35.45
    Cl_atomic_weight = 22.99

    return mass_in_grams / (Na_atomic_weight + Cl_atomic_weight)

weighted_mass_in_grams = 100
print("> {:.2f} grams of NaCl contains {:.2f} moles.".format(weighted_mass_in_grams, get_NaCl_mole(weighted_mass_in_grams)))
```
